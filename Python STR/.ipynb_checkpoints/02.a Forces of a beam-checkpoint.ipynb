```python
#Calculating the internal forces of a beam using NumPy

import numpy as np

l=5 #m
q=20 #kN/m

m=q*(l**2)/8
print("m =", m, "kN·m")
```

```python
#Calculating the bending moment and shear

x=np.linspace(0,l,20)

M=q/2*(l*x-x**2)
V=q*(l/2-x)
print(M)
print(V)
```

```python
#Matplotlib to display M and V

from matplotlib import pyplot as plt
plt.figure(figsize=(10,4))
plt.plot([0]*len(x), color='k')
plt.plot(-M)
plt.plot(V)
```
